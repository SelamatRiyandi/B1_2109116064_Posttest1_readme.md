## Algoritma Naive Bayes dan Preprocessing (pengenalan)

```python
# mengambil dataset df dari blok code di atas, menghapus kolom Unnamed: 0, CLIENTNUM, Customer_Age, dan Credit_Limit lalu menempatkannya pada variabel baru (dfnew)
dfnew = df.drop(columns = ["Unnamed: 0","CLIENTNUM","Customer_Age","Credit_Limit"])

# memuat kolom-kolom atribut pada data dfnew menjadi sebuah list dan dimuat ke dalam variabel atribut
atribut = list(dfnew.columns)
# menghapus lalu menambahkan kembali kolom 'Attrition_Flag' dengan tujuan untuk merubah posisi kolom tersebut agar berada di paling kanan (menyesuaikan dengan kebutuhan)
atribut.remove('Attrition_Flag')
atribut.append('Attrition_Flag')
# memuat kembali variabel list yang berisi kolom-kolom yang urutannya telah disesuaikan ke dalam variabel data dfnew
dfnew = dfnew[atribut]

dfnew
```

```python
import numpy as np

# membuat sebuah variabel (n_train) untuk menampung data jumlah/banyaknya data training dan data testing
n_train = len(dfnew)
# membuat sebuah variabel (target_counts) untuk menampung data jumlah kelas target
target_counts = dfnew.groupby('Attrition_Flag')['Attrition_Flag'].count()
# membuat sebuah variabel (prior_b)) untuk menampung operasi perhitungan probabilitas priori untuk setiap kelas target
prior_prob = target_counts / n_train
# menghitung probabilitas kondisional untuk setiap fitur pada setiap kelas target
cond_prob = {}
for col in dfnew.columns[:-1]:
    cond_prob[col] = {}
    for val in dfnew[col].unique():
        for target in dfnew['Attrition_Flag'].unique():
            count = len(dfnew[(dfnew[col] == val) & (dfnew['Attrition_Flag'] == target)])
            cond_prob[col][(val, target)] = count / target_counts[target]

# mendefinisikan fungsi klasifikasi dengan algoritma Naive Bayes
def naive_bayes(x):
    prob_yes = prior_prob['Attrited Customer']
    prob_no = prior_prob['Existing Customer']
    for i, col in enumerate(dfnew.columns[:-1]):
        try:
          prob_yes *= cond_prob[col][(x[i], 'Attrited Customer')]
          prob_no *= cond_prob[col][(x[i], 'Existing Customer')]
        except IndexError:
            return "Error: the input features are not compatible with the training data"
    if prob_no > prob_yes:
        return ('Existing Customer', prob_no)
    else:
        return ('Attrited Customer', prob_yes)

# uji coba fungsi naive_bayes
test = naive_bayes(["F",	"Uneducated","Single", "Unknown","Blue"])
print(test)
```
